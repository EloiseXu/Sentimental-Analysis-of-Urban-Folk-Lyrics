```python
import os
from random import choice
import re

import jieba.analyse
import jieba

import json
```

```python
lyrics = {}

invalid = ['封面设计', '出品', '：', '编曲', '词曲', '作曲', '作词', '编曲', '制作人', '监制', '混音', '合声', '平面美术设计',
          '出品', '封面手绘&设计', '录音', '母带', '和声', 'OP', ':', '\'', '.', '’', '／', '“', '”', '，', '-']

STOP_WORDS = set([w.strip() for w in open("stopwords.txt").readlines()])

X,Y = ['\u4e00','\u9fa5']
```

```python
root_dir = 'songlist_rock/'

list = os.listdir(root_dir)
for d in list:
    path = os.path.join(root_dir,d)
    lyric = ''
    with open(path, 'r') as f:
        lines = f.readlines()
        for line in lines:
            empty = False
            
            for x in invalid:
                if x in line:
                    empty = True
                    continue
            if empty:
                continue
                
            line = re.sub("[A-Za-z0-9\!\%\[\]\,\。\.\'\’\~]", "", line)
            if len(line.strip()) == 0:
                continue

            lyric = lyric + ' ' + line.strip() 
            
        if len(lyric.strip()) == 0:
            continue
                
        tag = jieba.cut(lyric,cut_all=False)
        tag = [ i for i in tag if len(i) >= 2 and X<=i<=Y and i not in STOP_WORDS ]
        output = ' '.join(tag)
                
        if len(lyric)>0:
            lyrics[d[:-4]]=output
```
